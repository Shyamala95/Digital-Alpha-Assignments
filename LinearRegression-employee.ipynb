```python
import pandas as pd
from sklearn.metrics import mean_squared_error
df=pd.read_csv(r'C:\Users\user\Desktop\employees.csv')
```

```python
#converting the non-numeric values to numerics 
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
le.fit(df.Department)
df['Department']=le.transform(df.Department)
```

```python
from sklearn import linear_model
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['Department','Worked Hours','Certification','Experience in years']],df.Salary,test_size=0.25,random_state=42)
```

```python
reg=linear_model.LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_predict=reg.predict(x_test)
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
rms=sqrt(mean_squared_error(y_test,y_predict))
print("Accuracy is ",rms )
```
